```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path='/content/drive/MyDrive/pretty little flower.jpg' # Replace with your image path
```

```python

def histogram_equalization(image_path):
    # Load image in grayscale
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Check if the image is loaded properly
    if img is None:
        raise ValueError(f"Error: Unable to load image at {image_path}. Check the file path and format.")

    # Apply histogram equalization using OpenCV
    equalized_img = cv2.equalizeHist(img)

    # Compute histograms
    hist_original, bins_original = np.histogram(img.flatten(), 256, [0, 256])
    hist_equalized, bins_equalized = np.histogram(equalized_img.flatten(), 256, [0, 256])

    # Display original and equalized images
    plt.figure(figsize=(20, 10))

    plt.subplot(2, 2, 1)
    plt.title('Original Image')
    plt.imshow(img, cmap='gray')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.title('Equalized Image')
    plt.imshow(equalized_img, cmap='gray')
    plt.axis('off')

    # Display histograms
    plt.subplot(2, 2, 3)
    plt.title('Original Histogram')
    plt.plot(hist_original, color='black')

    plt.subplot(2, 2, 4)
    plt.title('Equalized Histogram')
    plt.plot(hist_equalized, color='black')

    plt.tight_layout()
    plt.show()

    return equalized_img

```

```python
# Example usage
histogram_equalization(image_path)
```
